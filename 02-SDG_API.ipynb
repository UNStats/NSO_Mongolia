## Get Indicator Data:

```python
def indicator_data(goal, target, indicatorID, year):
    http = urllib3.PoolManager()
    response = http.request('GET', "http://sdg.gov.mn/SDG_WebAPI/DataAimagSoum/"+indicatorID+"/" + str(year))
    responseData = json.loads(response.data.decode('UTF-8'))
    
    x = responseData['Data']['ResponseData']['DataAimagSoum']
    
    response = http.request('GET', "http://sdg.gov.mn/SDG_WebAPI/VariableDataList/"+goal+"/G/EN")
    responseData = json.loads(response.data.decode('UTF-8'))
    
    g = responseData['Data']['ResponseData']['VariableList']
    
    goalID = g['GoalID']
    goalCode = g['Code']
    goalLabel = g['Label']
    
    for t in g['VariableList']:
        if t['Code'] == 'Target ' + target:
            targetID = t['Code']
            targetLabel = t['Label']
            
            for i in t['VariableList']:
                if i['Code'] == indicatorID:
                    indicatorLabel = i['Label']
                    break
    
    for row in x:
        row['goalID'] = goalID
        row['goalCode'] = goalCode
        row['goalLabel'] = goalLabel
        row['targetID'] = targetID
        row['targetLabel'] = targetLabel
        row['indicatorID'] = indicatorID
        row['indicatorLabel'] = indicatorLabel
        
    y = pd.DataFrame(x)
        
    
    return y
```

```python
x = indicator_data('9','9.2','9.2.2', 2015)

x.rename(index=str, columns={"AimagName": "REF_AREA_DESC", 
                              "aimagCode": "REF_AREA",
                              "data": "OBS_VALUE",
                              "yearCode": "TIME_PERIOD"})
    
print(goalID)
print(goalCode)
print(goalLabel)
print('---')
print(targetID)
print(targetLabel)
print('---')
print(indicatorID)
print(indicatorLabel)
```
